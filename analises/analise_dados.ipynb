```python
dados = 'dados_ficha_a_desafio.csv'

df = pd.read_csv(dados)

df.info()
```

```python
df
```

```python
# Função para corrigir a codificação, remover espaços e remover aspas, colchetes e caracteres específicos
def corrigir_limpar_e_strip(texto):
    if isinstance(texto, str):
        texto_corrigido = texto.encode('latin1').decode('unicode_escape')
        texto_limpo = texto_corrigido.replace('"', '').replace("'", '').replace('[', '').replace(']', '').strip()
        return texto_limpo
    return texto

df = df.map(corrigir_limpar_e_strip)

df
```

```python
colunas = ['sexo', 'obito', 'raca_cor', 'luz_eletrica',	'escolaridade',	'renda_familiar', 'em_situacao_de_rua', 'frequenta_escola',	'meios_transporte', 'doencas_condicoes', 'identidade_genero', 'meios_comunicacao', 'orientacao_sexual', 'possui_plano_saude', 'em_caso_doenca_procura', 'situacao_profissional', 'vulnerabilidade_social', 'data_atualizacao_cadastro',	'familia_beneficiaria_auxilio_brasil', 'crianca_matriculada_creche_pre_escola', 'tipo']

# Verificar valores únicos em cada coluna especificada
for coluna in colunas:
    valores_unicos = df[coluna].unique()
    print(f"Valores únicos na coluna '{coluna}':")
    print(valores_unicos)
```
